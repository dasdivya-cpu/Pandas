```python
l1=[101,102,103]
s3=pd.Series(l1,index=['A','B','C'])
#print(series3)
print(s3.index)
print(s3.values)
print(s3.dtype)
print(s3.shape)
print(s3.size)
print(s3.nbytes)
```

```python
#functions data analysis
l1=[101,102,103,np.NaN,101]
s3=pd.Series(l1,index=['A','B','C','D','E'])
print(s3.empty)
print(s3.hasnans)
print(len(s3))
print(s3.std())
print(pd.unique(s3))
print(s3.value_counts())
```
