Filling Missing Data in a Pandas DataFrame
Removing missing data also removes any associated data from those records.fillna() method can be applied to either an entire DataFrame or to a single column. By filling a constant in as a parameter, all missing values are replaced with that value:

```python
df3 = df1.fillna(0)
print(df3)
```

Fill with mean,median and mode

```python
x=df1['Age'].mean()
df4=df1['Age'].fillna(x)
print(df4)
```

```python
print(df1)
```

```python
x=df1['Age'].mean()
df1['Age'].fillna(x,inplace=True)
print(df1)
```

Filling with median

```python
df1 = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Evan', 'Kyra', np.NaN],
    'Age': [33, 32, 40, 57, np.NaN],
    'Location': ['Toronto', 'London', 'New York', np.NaN, np.NaN]
})
df1['Age'].fillna(df1['Age'].median(),inplace=True)
print(df1)
```

Filling with mode

```python
df1 = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Evan', 'Kyra', np.NaN],
    'Age': [33, 32, 40, 57, np.NaN],
    'Location': ['Toronto', 'London', 'New York', np.NaN, np.NaN]
})
df1['Age'].fillna(df1['Age'].mode()[0],inplace=True)
print(df1)
```

Replace method

```python
df1 = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Evan', 'Kyra', np.NaN],
    'Age': [33, 32, 40, 57, np.NaN],
    'Location': ['Toronto', 'London', 'New York', np.NaN, np.NaN]
})
df1.replace(to_replace = np.NaN, value =-99999,inplace=True)
print(df1)
```

Removing Duplicate values
DataFrame.duplicated(subset=None, keep='first')[source]
Return boolean Series denoting duplicate rows.

Considering certain columns is optional.

Parameters
subsetcolumn label or sequence of labels, optional
Only consider certain columns for identifying duplicates, by default use all of the columns.

keep{‘first’, ‘last’, False}, default ‘first’
Determines which duplicates (if any) to mark.

first : Mark duplicates as True except for the first occurrence.

last : Mark duplicates as True except for the last occurrence.

False : Mark all duplicates as True.

Returns
Series
Boolean series for each duplicated rows.

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df.duplicated()
```

By using ‘last’, the last occurrence of each set of duplicated values is set on False and all others on True.

```python
df.duplicated(keep='last')
```

To find duplicates on specific column(s),

```python
df.duplicated(subset=['brand'])
```

DataFrame.drop_duplicates(subset=None, *, keep='first', inplace=False, ignore_index=False)[source]
Return DataFrame with duplicate rows removed.

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df
```

By default, it removes duplicate rows based on all columns.

```python
df.drop_duplicates()
```

To remove duplicates on specific column(s), use subset.

```python
df.drop_duplicates(subset=['brand'])
```

To remove duplicates and keep last occurrences, use keep.

```python
df.drop_duplicates(subset=['brand', 'style'], keep='last')
```
