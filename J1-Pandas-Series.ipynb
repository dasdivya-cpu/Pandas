The Pandas Series can be defined as a one-dimensional array that is capable of storing various data types. We can easily convert the list, tuple, and dictionary into series using "series' method. The row labels of series are called the index. 

Series by using various inputs:

    Array

```python
import pandas as pd  
import numpy as np  
info = np.array(['P','a','n','d','a','s'])  
a = pd.Series(info)  
print(a)   
```

Series by using various inputs:

  
    Dict

```python
info1 = {'x' : 0., 'y' : 1., 'z' : 2.}  
a = pd.Series(info1)  
print (a)  
```

Series by using various inputs:

   
    Scalar value


```python
x = pd.Series(4, index=[0, 1, 2, 3])  
print (x)  
```

Accessing data from series with Position:

```python
x = pd.Series([1,2,3],index = ['a','b','c'])  
#retrieve the first element  
print (x[0])  
```

Series object attributes

The Series attribute is defined as any information related to the Series object such as size, datatype. etc. 

```python
x=pd.Series(data=[2,4,6,8])   
y=pd.Series(data=[11.2,18.6,22.5], index=['a','b','c'])   
print(x.index)   
print(x.values)   
print(y.index)   
print(y.values)  
print(x.dtype)   
print(x.shape)    
print(y.shape)  
print(x.ndim, y.ndim)   
print(x.size, y.size)   
print(x.nbytes, y.nbytes)  
```

Checking Emptiness and Presence of NaNs

```python
a=pd.Series(data=[1,2,3,np.NaN])   
b=pd.Series(data=[4.9,8.2,5.6],index=['x','y','z'])   
c=pd.Series()   
print(a.empty,b.empty,c.empty)   
print(a.hasnans,b.hasnans,c.hasnans)   
print(len(a),len(b))   
print(a.count( ),b.count( ))  
```

**Series Functions**

Pandas Series.map(): is used to map the values from two series that have a common column. To map the two Series, the last column of the first Series should be the same as the index column of the second series, and the values should be unique.

```python
a = pd.Series(['Java', 'C', 'C++', np.nan])  
b=a.map({'Java': 'Python'})
a 
b 
```

Pandas Series.std()

The Pandas std() is defined as a function for calculating the standard deviation of the given set of numbers, DataFrame, column, and rows.

```python
import numpy as np   
c=pd.Series([4,7,2,1,6,3])
print(c.std())  
```

Pandas Series.unique()

While working with the DataFrame in Pandas, you need to find the unique elements present in the column. For doing this, we have to use the unique() method to extract the unique values from the columns. 

```python
pd.unique(pd.Series([2, 1, 3, 3]))  
```

Pandas Series.value_counts()

The value_counts() function returns a Series that contain counts of unique values. It returns an object that will be in descending order so that its first element will be the most frequently-occurred element.

```python
index = pd.Index([2, 1, 1, np.nan, 3])  
index.value_counts()  
```

Pandas Series.to_frame():
Series is defined as a type of list that can hold an integer, string, double values, etc. It returns an object in the form of a list that has an index starting from 0 to n where n represents the length of values in Series.

```python
emp = ['Parker', 'John', 'Smith', 'William']   
id = [102, 107, 109, 114]     
emp_series = pd.Series(emp)   
id_series = pd.Series(id)     
frame = { 'Emp': emp_series, 'ID': id_series }   
result = pd.DataFrame(frame)     
print(result)  
```
